```python
import pandas as pd
from datetime import datetime
from dateutil.parser import parse
from matplotlib import pyplot as plt
import scipy.stats as stats
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('Bitfinex_BTCEUR_trades_2018_02_02.csv')
df
```

```python
pd.to_datetime(df['date'])

```

```python
amountdata= sorted(df['amount'])
std = np.std(amountdata)
mean = np.mean(amountdata)
pdf = stats.norm.pdf(amountdata, mean, std)
plt.title('Gaussian distribution curve/ fisher tranform for Amounts of Bitfex')
plt.plot(amountdata, pdf)

```

```python
pricedata= sorted(df['price'])
std = np.std(pricedata)
mean = np.mean(pricedata)
pdf = stats.norm.pdf(pricedata, mean, std)
plt.plot(pricedata, pdf)
```
